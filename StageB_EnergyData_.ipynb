**Now that we left with relevant columns, we will move to the next level which normalisation**  
Normalization is a rescaling of the data from the original range so that all values are within the new range of 0 and 1.

  y = (x – min) / (max – min)


```python
#we will use MinMaxScaler for the Normalisation
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dfe_Nich_normalised = pd.DataFrame(scaler.fit_transform(dfe_Nich), columns=dfe_Nich.columns)
```

**The above result is good. Next is seperating the Target variable('Appliances') from Predicted variables. Using Drop method**

```python
#Target is Appliances while the predictors are Temperature (T1-T9), Room_Humidity(RH1-RH9) and the two random variables etc
dfe_features = dfe_Nich_normalised.drop(columns=['Appliances'])
appliances_target = dfe_Nich_normalised['Appliances']
```

At this juncture we start to **Build the Model** first by spliting train-test(70-30) with a random state of 42

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dfe_features, appliances_target, test_size=0.3, random_state=42)
```

# **Linear Regretion**
This is the first part of our modeling

```python
#Import Linear Regression
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
# Fitting the model to the training data
linear_model.fit(x_train, y_train)
```

```python
# Predict the values
predicted_values = linear_model.predict(x_test)
```

 **From SKlearn libraries we import 'Matrics" for proper evaluation**

```python
# Import Metrics for evaluation
from sklearn import metrics
```

```python
# To Publish the Metrics for Evaluation we use the method below.
print('Mean Absolute Error (MAE)', metrics.mean_absolute_error(y_test,predicted_values))
print('Mean Square Error (MSE)', metrics.mean_squared_error(y_test,predicted_values))
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,predicted_values)))
print('R2 score', metrics.r2_score(y_test,predicted_values))
```

#**PENALIZATION METHOD**

1.   Ridge Regression
2.   Lasso regresion

 **Ridge Regression**

```python
#Using Ridge Regression with Alpha = 0.5
from sklearn.linear_model import Ridge
ridge_reg1 = Ridge(alpha=0.5)
ridge_reg1.fit(x_train, y_train)
pred1 = ridge_reg1.predict(x_test)           # test data is predicted with this model
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print('R2 score', metrics.r2_score(y_test,pred1))
```

```python
#Using Ridge Regression with Alpha = 1.5
ridge_reg2 = Ridge(alpha=1.5)
ridge_reg2.fit(x_train, y_train)
pred2 = ridge_reg2.predict(x_test) # test data is predicted with this model
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,pred2)))
print('R2 score', metrics.r2_score(y_test,pred2))
```

```python
#Using Ridge Regression with Alpha = 2
ridge_reg3 = Ridge(alpha=2)
ridge_reg3.fit(x_train, y_train)
pred2 = ridge_reg3.predict(x_test) # test data is predicted with this model
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,pred2))) #RMSE
print('R2 score', metrics.r2_score(y_test,pred2)) #r2
```

```python
#Using Ridge Regression with Alpha = 5
ridge_reg4 = Ridge(alpha=5)
ridge_reg4.fit(x_train, y_train)
pred2 = ridge_reg4.predict(x_test)  # test data is predicted with this model
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,pred2))) #(RMSE)
print('R2 score', metrics.r2_score(y_test,pred2)) #R2
```

**LASSO REGRESSION**

Also known as L2 Regularisation, this is a technique that uses a penalty term to shrink the magnitude of coefficients towards zero without eliminating them.

```python
#Using Lasso Regression with Alpha = 0.001
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
pred5 = lasso_reg.predict(x_test)
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,pred5)))
print('R2 score', metrics.r2_score(y_test,pred5))
```

```python
#Using Lasso Regression with Alpha = 0.1
lasso_reg1 = Lasso(alpha=0.1)
lasso_reg1.fit(x_train, y_train)
pred6 = lasso_reg1.predict(x_test)
print('Root Mean Square Error (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,pred6)))
print('R2 score', metrics.r2_score(y_test,pred6))
```

```python
# Comparing the Effects of Regularisation
def get_weights_df(model, feat, col_name):
  #this function will return the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'linear_model_Weight')
ridge_weights_df = get_weights_df(ridge_reg1, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
